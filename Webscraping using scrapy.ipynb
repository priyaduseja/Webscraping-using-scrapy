##### Scraping Quotes to scrap website: Extract the tags along with the quotes and author name

```python
import urllib3
urllib3.disable_warnings(urllib3.exceptions.InsecureRequestWarning)
import scrapy
from scrapy.crawler import CrawlerProcess

class ScratchQuotes(scrapy.Spider):
    name = 'quotes'
    start_urls = ['https://quotes.toscrape.com/']
    
    def parse(self, response):
        for div in response.css('.quote'):
            quote = div.css('.text::text').get()
            author = div.css('.author::text').get()
            tags = div.css('.tag::text').getall()
            tags = ' , '.join(tags)
            yield { 'Quote': quote.replace('“','').replace('”',''), 'Author': author, 'Tags': tags}
            
process = CrawlerProcess(settings = {'FEEDS': {'Quotes_from_scrapy.csv': {'format': 'csv'}}})
process.crawl(ScratchQuotes)
process.start() 
```
